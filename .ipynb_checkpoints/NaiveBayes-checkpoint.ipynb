```python
import numpy as np
import matplotlib.pyplot as plt
from util import DataGenerator, DataPlotter
np.random.seed(42)
```

```python
num_samples = 50
num_features = 2
type = "classification"
if type == "classification":
    num_classes = 2

gen = DataGenerator(num_samples, num_features, num_classes)
X, y = gen.get_data()
plotter = DataPlotter()
plotter.plot_data(X,y, type, num_classes)

num_samples, num_features = X.shape
unique_classes, class_counts = np.unique(y, return_counts=True)
class_counts
```

```python
# Calculate Prior Probabilities
prior_probabilities = class_counts / num_samples
print("Classes:", unique_classes)
print("Prior Probabilities:", prior_probabilities)

# Calculate Likelihoods
likelihoods = {}
for cls in unique_classes:
    X_for_class = X[y == cls].flatten()
    mean = np.mean(X_for_class)
    std_dev = np.std(X_for_class)
    class_likelihoods = 1 / (np.sqrt(2 * np.pi) * std_dev) * np.exp(-0.5 * ((X - mean) / std_dev) ** 2)
    likelihoods[cls] = class_likelihoods

for cls, likelihood in likelihoods.items():
    print(f"Class {cls} Likelihoods:", likelihood)
# Calculate Posterior Probabilities
```
