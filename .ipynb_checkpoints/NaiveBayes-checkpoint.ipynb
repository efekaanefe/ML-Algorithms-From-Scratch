```python
import numpy as np
import matplotlib.pyplot as plt
from util import DataGenerator, DataPlotter
np.random.seed(42)
```

```python
num_samples = 10
num_features = 2
type = "classification"
if type == "classification":
    num_classes = 2

gen = DataGenerator(num_samples, num_features, num_classes)
X, y = gen.get_data()
plotter = DataPlotter()
plotter.plot_data(X,y, type, num_classes)
```

```python
# X = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
# y = np.array([1, 1, 0, 0])
```

```python
# Gaussian naive Bayes
num_samples, num_features = X.shape
classes, class_counts = np.unique(y, return_counts=True)
num_classes = len(classes)

priors = class_counts / num_samples
print("Priors:", priors)
print("-"*50)

for idx, cls in enumerate(classes):
    X_for_class = X[y == cls]
    mean_for_class = np.mean(X_for_class, axis=0) # for the given class, mean values of the features
    std_for_class = np.std(X_for_class, axis=0) # for the given class, std values of the features
    print(f"Mean and std values of the features for class=={cls} ")
    print("Mean:", mean_for_class)
    print("Std:", std_for_class)
    print("-"*50)
    
    
```
