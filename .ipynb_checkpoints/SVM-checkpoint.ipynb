```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.random.rand(100, 2)  # 100 samples with 2 features
y = np.random.randint(0, 2, size=100)  # Binary class labels

# Fit Naive Bayes classifier (not necessary for visualization, just for example)
# Assuming you have weights for each feature
weights = np.array([0.5, -0.3])  # Example weights
bias = 0  # Example bias (not used in this visualization)
hyperplane_slope = -weights[0] / weights[1]
hyperplane_intercept = -bias / weights[1]

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')

# Plot decision boundary (hyperplane)
x_values = np.linspace(0, 1, 10)
y_values = hyperplane_slope * x_values + hyperplane_intercept
plt.plot(x_values, y_values, color='k', linestyle='--')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Naive Bayes Decision Boundary')
plt.show()
```
