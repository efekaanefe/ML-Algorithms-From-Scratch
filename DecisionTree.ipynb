## Using CART algorithm

```python
class Node:
    def __init__(self, feature_index=None, threshold=None, value=None, left=None, right=None):
        self.feature_index = feature_index  # Index of the feature to split on
        self.threshold = threshold  # Threshold value for splitting numerical features
        self.value = value  # Prediction value if the node is a leaf node
        self.left = left  # Left child node
        self.right = right  # Right child node
```

```python
class DecisionTree:
    def __init__(self, max_depth=4, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split

    def fit(self, X, y):
        self.tree = self._build_tree(X, y, depth=0)

    def _build_tree(self, X, y, depth):
        if self._is_leaf_node(X, y, depth):
            return Node(value=self._find_majority_class(y))

        # Find the best split for the current node
        best_feature_index, best_threshold = self._find_best_split(X, y)
        
        # Split the data based on the best split
        X_left, y_left, X_right, y_right = self._split_data(X, y, best_feature_index, best_threshold)
        
        # Recursively build the left and right subtrees
        left_subtree = self._build_tree(X_left, y_left, depth + 1)
        right_subtree = self._build_tree(X_right, y_right, depth + 1)
        
        # Create a node to represent the best split
        return Node(feature_index=best_feature_index, threshold=best_threshold, 
                    left=left_subtree, right=right_subtree)

    def _is_leaf_node(self, X, y, depth):
        # Check if stopping criteria are met
        if depth >= self.max_depth: return True
        if len(y) < self.min_samples_split: return True
        if len(np.unique(y)) == 1: return True
        return False

    def _find_majority_class(self, y):
        # Find the majority class in the target labels
        unique_classes, counts = np.unique(y, return_counts=True)
        majority_class_index = np.argmax(counts)
        return unique_classes[majority_class_index]

    def _split_data(self, X, y, best_feature_index, best_threshold):
        feature_values = X[:, best_feature_index]
        
        X_left = X[feature_values <= best_threshold]
        y_left = y[feature_values <= best_threshold]
        X_right = X[feature_values > best_threshold]
        y_right = y[feature_values > best_threshold]
        
        return X_left, y_left, X_right, y_right


    def _find_best_split(self, X, y):
        best_gini = np.inf
        best_feature_index = None
        best_threshold = None
        
        # Iterate over each feature
        for feature_index in range(X.shape[1]):
            # Sort the feature values to find potential split points
            feature_values = np.sort(np.unique(X[:, feature_index]))
            
            # Iterate over potential split points
            for threshold in feature_values:
                # Split the data
                X_left, y_left, X_right, y_right = self._split_data(X, y, feature_index, threshold)
                
                # Calculate Gini impurity for the split
                gini_left = self._gini_impurity(y_left)
                gini_right = self._gini_impurity(y_right)
                gini = (len(y_left) * gini_left + len(y_right) * gini_right) / len(y)
                
                # Update the best split if this split is better
                if gini < best_gini:
                    best_gini = gini
                    best_feature_index = feature_index
                    best_threshold = threshold
        
        return best_feature_index, best_threshold

    def _gini_impurity(self, y):
        classes = np.unique(y)
        impurity = 1
        for cls in classes:
            p_cls = np.mean(y == cls)
            impurity -= p_cls ** 2
        return impurity



    def predict(self, X):
        predictions = np.empty(len(X), dtype=np.intc)
        # Traverse the decision tree for each sample in X
        for i, sample in enumerate(X):
            node = self.tree  # Start at the root node
            # Traverse down the tree until a leaf node is reached
            while node.left:
                if sample[node.feature_index] <= node.threshold: node = node.left
                else: node = node.right
            # Assign the predicted class of the leaf node to the corresponding index
            predictions[i] = node.value
        return predictions

```

```python
tree = DecisionTree(max_depth=5, min_samples_split=5)
tree.fit(X, y)
predictions = tree.predict(X)
accuracy = np.mean(predictions == y)
print("Accuracy:", accuracy)
```
